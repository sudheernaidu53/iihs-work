## Output s and their formats:
nic_class.csv has two columns : 'class' and 'details' both are strings, 'class' is a 4 digit nic code

nic_subclass.csv has two colmns : 'subclass' and 'details', both of them are strings, 'subclass' is a 5 digit nic code

nic_group.csv has two columns: 'group' and 'details', both are strings, 'group' is 3 digit nic code

nic_division.csv  has three columns: 'division', 'details' and 'name', all of them are strings. 'division' is 2 digit nic code. and 'details' is the description of all groups,classes,subclasses under that division. 'name is the 'name' of division 

nic_section.csv has three columns: 'section', 'details' and 'name', all of them are strings. 'section' is 1 alphabet nic code. and 'details' is the description of all divisions,groups,classes,subclasses under that division. 'name' is the name of section.

## Details about the NIC document
All the codes are expressed under the classes 'group','class','subclass,

subclass has code with 5 letters

class has code with 4 letters

group has code with 3 letters

### detaching nic codes which are mixed up with description

some of the rows has no entries under group,class or subclass but all those nic codes are mixed up with details column.
I have manually seperated classes and groups as they are less in number, following code detaches those mix-ups and put them in the right place

```python
def numberpartitioner(k):
    if (type(k[0])==type('a')):
#         print(k)
#         print(k[0],"considered as string \n")
        return k
        pass
    else:
        try :
            k[0]= int(k[1].split(' ',1)[0])
            k[1] = k[1].split(' ',1)[1]
            return k
        except:
            return k
            pass
```

```python
fakedf = (sampledf[["subclass","details"]][:].T.apply(numberpartitioner)).T
sampledf[["subclass","details"]]=fakedf
sampledf
```

there are false entries in subclass with 4 integers or 3 integers in subclass and 3 integers in class

#### subclass 
when group is 11, 1111 -  is correct entry

where as when group is 220 , 220 or 2200 is incorrect.

##### class
when group is 11, 111 is correct entry

where as when group is 220, 220 is false entry.

solution is to exploit that there should not be any codes in class or subclass when there are codes in groups.

and we don't need to care about Nan because we can't compare nans with each other

|||ly no codes in subclass if there is in class.

```python
for row in range(nic_df.shape[0]):
#     if (type(nic_df["subclass"][row])==type('a')) and (type(nic_df["subclass"][row])==type('a'))
    if (nic_df["subclass"][row] == nic_df["class"][row]) or (nic_df["subclass"][row] == nic_df["group"][row]):
        nic_df["subclass"][row] = np.nan
    if (nic_df["class"][row] == nic_df["group"][row]):
        nic_df["class"][row] = np.nan
# nic_df['group'][0]=np.nan

nic_df

#check that earlier on 4597th row, 990 was in both group and subclass, now subclass thing has been turned to NaN
```

```python
# print(nic_df.columns)
# nic_df.drop(['Column'],axis=1,inplace=True)
```

```python
for row in range(nic_df.shape[0]):
    if (type(nic_df["subclass"][row])==type('a')) and (len(nic_df["subclass"][row]))<5:
#         print(row)
        nic_df["subclass"][row]=(nic_df["subclass"][row]).zfill(5)
    if (type(nic_df["class"][row])==type('a')) and (len(nic_df["class"][row]))<4:
        nic_df["class"][row]=(nic_df["class"][row]).zfill(4)
    if (type(nic_df["group"][row])==type('a')) and (len(nic_df["group"][row]))<3:
        nic_df["group"][row]=(nic_df["group"][row]).zfill(3)   
```
