```python
#count number of 2 digit, 3 digit codes, 4 digit codes..... in nic_codes
tmpdf7 = df7[['company_name','nic_prod_code','nic_name']].copy()
def lenone(integer):
    if (len(str(integer))) == 1:
        return 1
    return 0
def lentwo(integer):
    if (len(str(integer))) == 2:
        return 1
    return 0
def lenthree(integer):
    if (len(str(integer))) == 3:
        return 1
    return 0
def lenfour(integer):
    if (len(str(integer))) == 4:
        return 1
    return 0
def lenfive(integer):
    if (len(str(integer))) == 5:
        return 1
    return 0
```

```python
def divideinto(number):
    if (nic_class['class'] == number).sum():
        return str(number).zfill(4)
    if (nic_group['group'] == number).sum():
        return str(number).zfill(3)
    if (nic_subclass['subclass'] == number).sum():
        return str(number).zfill(5)
    else:
        return number
tmpdf7['correctlength'] = tmpdf7['nic_prod_code'].apply(divideinto)
```
