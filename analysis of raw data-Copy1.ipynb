```python
def toint(string):
    return(int(string))
def setthem(array):
    return list(set(array))
def divideinto(number):
    if (nic_class['class'] == number).sum():
        return str(number).zfill(4)
    if (nic_group['group'] == number).sum():
        return str(number).zfill(3)
    if (nic_subclass['subclass'] == number).sum():
        return str(number).zfill(5)
    else:
        return number
# def apply_or(arr):
# #     if len(arr)==3:
#     arr = list(arr)
#     return (arr[0] or arr[1])
# #     else
def stripto4(a):
    if len(str(a))==5:
        return(str(a)[:4])
    else:
        return str(a)
def stripto3(a):
    if len(str(a))>3:
        return(str(a)[:3])
    else:
        return str(a)
def stripto2(a):
    if len(str(a))>2:
        return(str(a)[:2])
    else:
        return str(a)
def makeitstring(arr,sep = ' '):
    return sep.join(arr)
```

```python
def lenone(integer):
    if (len(str(integer))) >= 1:
        return 1
    return 0
def lentwo(integer):
    if (len(str(integer))) >= 2:
        return 1
    return 0
def lenthree(integer):
    if (len(str(integer))) >= 3:
        return 1
    return 0
def lenfour(integer):
    if (len(str(integer))) >= 4:
        return 1
    return 0
def lenfive(integer):
    if (len(str(integer))) >= 5:
        return 1
    return 0
```

```python
def masterdf(address = 0,nic_type = 'group',extra = None,mergeon = 'company_name',groupon = 'company_name',file = None,*args,**kwargs):
    if file == None:
        return None
    if len(args):
        for arg in args:
            file = pd.merge(file,arg,how='outer',on = mergeon)
    colstokeep = ['company_name','nic_prod_code','nic_name'] + extra #extra is a list of column names
    if address:
        colstokeep = colstokeep+['address']
    file = file[[colstokeep]]    
    file['correctlength'] = file['nic_prod_code'].apply(divideinto)
    
    if nic_type == 'subclass':
        file['filter'] = file['correctlength'].apply(lenfive)
    elif nic_type == 'class':
        file['filter'] = file['correctlength'].apply(lenfour)
        file['nic_prod_code'] = file['correctlength'].apply(stripto4)
    elif nic_type == 'group'
        file['filter'] = file['correctlength'].apply(lenthree)
        file['nic_prod_code'] = file['correctlength'].apply(stripto3)
    elif nic_type == 'division'
        file['filter'] = file['correctlength'].apply(lentwo)
        file['nic_prod_code'] = file['correctlength'].apply(stripto2)
        
    file = file[[colstokeep]][file['filter']]
    file = file.groupby(groupon).agg(lambda x:list(x))
    tempcols = colstokeep
    tempcols.remove(groupon)
    for i in tempcols:
        file[i] = file[i].apply(setthem)
```
